```python
import plotly.graph_objects as go

def plot_3d_line(datasets):
    fig = go.Figure()

    for idx, (data, color, name) in enumerate(datasets):
        x, y, z = zip(*data)  # Unpack points
        fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='lines+markers',
                                   line=dict(color=color, width=4),
                                   marker=dict(size=6),
                                   name=name))

    fig.update_layout(scene=dict(xaxis_title='n', yaxis_title='noofthreads', zaxis_title='timetaken'),
                      title='Interactive 3D Line Plot')
    fig.show()

# Plot with different colors
plot_3d_line([
    (dataset1, 'blue', 'Dataset 1'),
    (dataset2, 'red', 'Dataset 2'),
    (dataset3, 'green', 'Dataset 3'),
    (dataset4, 'purple', 'Dataset 4'),

])
```
